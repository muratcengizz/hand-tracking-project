```python
import cv2
import time
import mediapipe as mp
```

```python
cap = cv2.VideoCapture(0)

mpHand = mp.solutions.hands
hands = mpHand.Hands()

mpDraw = mp.solutions.drawing_utils

cTime, pTime = (0, 0)
while True:
    ret, img = cap.read()
    
    if not ret: break
    if cv2.waitKey(delay=1) == ord('q'): break
    
    results = hands.process(img).multi_hand_landmarks
    
    if results:
        for handLms in results:
            mpDraw.draw_landmarks(image=img, 
                                  landmark_list=handLms, 
                                  connections=mpHand.HAND_CONNECTIONS)
            for id, landmark in enumerate(handLms.landmark):
                height, width, color = img.shape
                cx, cy = int(landmark.x * width), int(landmark.y * height)
                
                #bilek
                """
                if id == 0:
                    cv2.circle(img=img, 
                               center=(cx, cy), 
                               radius=9, 
                               color=(255, 0, 0), 
                               thickness=cv2.FILLED) 
                if id == 20:
                    cv2.circle(img=img,
                               center=(cx, cy),
                               radius=8,
                               color=(0, 0, 255),
                               thickness=cv2.FILLED)
                """
    
    
                
    img = cv2.flip(src=img, flipCode=1)
    
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    
    img = cv2.putText(img=img, 
                      text=f"FPS: {round(fps, 2)}", 
                      org=(30, 40),
                      fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                      fontScale=1,
                      color=(0, 0, 255),
                      thickness=3)
    
    cv2.imshow(winname='Camera', mat=img)
    
    
    
    
    
cap.release()
cv2.destroyAllWindows()
```
